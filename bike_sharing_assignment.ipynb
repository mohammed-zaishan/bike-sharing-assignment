```python
# Removing 'instant' column as duplicate of index of dataframe and no significance to output
boom_bikes_df.drop(['instant'],axis=1,inplace=True)
# Removing 'casual' and 'registered' columns as they are reduntant, similar to cnt column
boom_bikes_df.drop(['casual','registered'],axis=1,inplace=True)
# First 5 rows of the data
boom_bikes_df.head()
```

```python
boom_bikes_df.dtypes
```

```python
# 'dteday' column
## Convert 'dteday' column to datetime format
boom_bikes_df['dteday'] = pd.to_datetime(boom_bikes_df['dteday'],dayfirst=True)
## Deriving "days since the data started"
boom_bikes_df['days_from_start'] = (boom_bikes_df.dteday - pd.to_datetime('2018-01-01',format='%Y-%m-%d')).dt.days
## Removing 'dteday' column
boom_bikes_df.drop(['dteday'],axis=1,inplace=True)
```

```python
# 'season' column
## Converting 'season' column to categorical
boom_bikes_df.season = boom_bikes_df.season.astype('str').map({'1':'spring', '2':'summer', '3':'fall', '4':'winter'})
```

```python
# 'mnth' column
## Converting 'mnth' column to string name of the month
boom_bikes_df.mnth = boom_bikes_df.mnth.astype('str').map({'1':'Jan', '2':'Feb', '3':'Mar', '4':'Apr', '5':'May', '6':'Jun', '7':'Jul', '8':'Aug', '9':'Sep', '10':'Oct', '11':'Nov', '12':'Dec'})
```

```python
# 'weekday' column
## Converting 'weekday' column to string name of the day
boom_bikes_df.weekday = boom_bikes_df.weekday.astype('str').map({'1':'Mon', '2':'Tue', '3':'Wed', '4':'Thu', '5':'Fri', '6':'Sat', '0':'Sun'})

## Deriving 'weekend' column
boom_bikes_df['weekend'] = boom_bikes_df.weekday.map({'Mon':0, 'Tue':0, 'Wed':0, 'Thu':0, 'Fri':0, 'Sat':1, 'Sun':1})
```

```python
# 'weathersit' column
## Converting 'weathersit' column to sting name of the weather situation
boom_bikes_df.weathersit = boom_bikes_df.weathersit.astype('str').map({'1':'Clear', '2':'Cloudy', '3':'Rainy', '4':'Storm'})
```

```python
# 'temp','atemp','hum','windspeed' columns
## Rounding the values of 'temp','atemp','hum','windspeed' columns to 3 decimal places
boom_bikes_df[['temp','atemp','hum','windspeed']] = boom_bikes_df[['temp','atemp','hum','windspeed']].round(3)
```

### Validating the assumptions of Linear Regression

Linear Relationship
<br>Homoscedasticity
<br>Absence of Multicollinearity
<br>Independence of residuals (absence of auto-correlation)
<br>Residuals are normally distributed

```python
# Confirming linear relationship 
sm.graphics.plot_ccpr(lm, 'temp')
plt.show()
```

The CCPR plot enables us to evaluate the influence of a single regressor on the response variable while taking into consideration the impacts of the other independent variables. It demonstrates that the connection between the independent factors and the response variable is linear.

```python
# Confirming linear Homoscedasticity
p = sns.scatterplot(x=y_train_pred,y=residuals)
plt.xlabel('y_pred/predicted values')
plt.ylabel('Residuals')
p = sns.lineplot(x=[min(y_train_pred), max(y_train_pred)],y=[0,0],color='blue')
p = plt.title('Residuals vs fitted values plot for homoscedasticity check')
plt.show()
```

#### Homoscedasticity is present since the variance of the residuals are almost constant.

```python
# Confirming Multi Colinearity
plt.figure(figsize=(15,8))
sns.heatmap(X_train_const.drop('const',axis=1).corr(),annot=True, cmap='YlGnBu')
plt.show()
```

#### As you can see independent variables are not highly correlated with each other.

```python
print('The Durbin-Watson value for Final Model is',round(sm.stats.stattools.durbin_watson((y_train - y_train_pred)),4))
```

The Durbin-Watson statistic ranges from 0 to 4, with a value close to 2 indicating non-autocorrelation, a value near 0 suggesting positive autocorrelation, and a value near 4 indicating negative autocorrelation. In this case, the statistic is extremely close to 2, which confirms non-autocorrelation.

```python
# Checking for normal distribution of residuals
fig = plt.figure()
sns.distplot((y_train - y_train_pred))
fig.suptitle('Error Terms')                  
plt.xlabel('Errors')     
plt.show()
```

#### The residuals are normally distributed thus we can say that the model follows all assumptions.
